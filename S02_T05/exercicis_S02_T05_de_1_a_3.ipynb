```python
import pandas as pd
import numpy as np

p = pd.read_csv('DelayedFlights.csv')
print(p.head(5))

df = p[["DepTime", "CRSDepTime", "ArrTime", "CRSArrTime", "ArrDelay", "UniqueCarrier", "FlightNum", "Origin", "Dest", "DepDelay", "Distance", "Cancelled"]]
df
```

```python
print("Resumen estadístic")
print(df.describe())
print()

for column in df:
    if df[column].isnull().values.any():
        print ("la columna", column, "té dades que falten")
print()

print("Velocitat mitjana del trajecte")
df["velocitat"] = df["Distance"] / (df["CRSArrTime"] - df["CRSDepTime"])
print(df["velocitat"])
print()

print("Itinerari")
df["Ori_Dest"] = df["Origin"] + "-" + df["Dest"]
print(df["Ori_Dest"])
print()

aeroliretrasos = df.groupby("UniqueCarrier")["ArrDelay"].sum().sort_values(ascending=False)
print("Llista de les 10 aerolínies amb més retràs acumulat (en minuts)")
print(aeroliretrasos.head(10))
print()
print()

df.set_index("Ori_Dest")
velocitatsmitjanas = df.groupby("Ori_Dest")["velocitat"].mean()
print("Les velocitats mitjanas dels trajectes en miles/min són")
print(velocitatsmitjanas)
print()

print("Els trajectes més llargs són")
print(df.groupby("Ori_Dest")["Distance"].mean().sort_values(ascending=False).head(10))
print()

df.reset_index()
```

```python
df.to_csv("noudelayedfligths.csv", index=False)
```
