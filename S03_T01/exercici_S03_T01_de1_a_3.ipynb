```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

p = pd.read_csv('DelayedFlights.csv')
df = p[["DepTime", "CRSDepTime", "ArrTime", "CRSArrTime", "ArrDelay", "UniqueCarrier", "FlightNum", "Origin", "Dest", "DepDelay", "AirTime", "Distance", "Cancelled"]]

df["UniqueCarrier"].value_counts().plot()
plt.suptitle("Vols per Aerolnia")
plt.xticks(np.arange(len(df["UniqueCarrier"].unique())), df["UniqueCarrier"].unique())
plt.show()


df["ArrDelay"].plot()
plt.xlabel("Minuts")  
plt.suptitle("Freqncia del temps de retard a l'arribada")
plt.show()


(df.groupby("UniqueCarrier")["ArrDelay"].sum().sort_values()).plot(kind="barh")
plt.xlabel("Minuts") 
plt.ylabel("") 
plt.suptitle("Retard acumulat a l'arribada per Aerolnia")
plt.show()


df.plot(kind='scatter', x = "ArrDelay", y = "DepDelay",color='red')
plt.suptitle("Correlaci entre els retards de l'arribada i la sortida de cada vol")
plt.show()


fig, ax1 = plt.subplots()
ax1.plot(df["UniqueCarrier"].unique(), df.groupby("UniqueCarrier")["ArrDelay"].sum(), label = "Arribades")
ax1.tick_params(axis='y', labelcolor = "b")
ax2 = ax1.twinx()
ax2.plot(df["UniqueCarrier"].unique(), df.groupby("UniqueCarrier")["DepDelay"].sum(), color='r', label="Sortides")
ax2.tick_params(axis='y', labelcolor = "r")
plt.xticks(np.arange(len(df["UniqueCarrier"].unique())), df["UniqueCarrier"].unique())
plt.legend()
plt.suptitle("Temps acumulat de retard a la sortida i a l'arribada per Aerolnia")
plt.show()


(df.groupby("UniqueCarrier")["ArrDelay"].sum().sort_values()).plot(kind= "line", y='ArrDelay', color='b', label="Retrs Arribada")
(df.groupby("UniqueCarrier")["DepDelay"].sum().sort_values()).plot(kind = "line", y='DepDelay', color='g', label="Retrs Sortida")
(df.groupby("UniqueCarrier")["AirTime"].sum().sort_values()).plot(kind = "line", y='AirTime', color='r', label="Temps del vol")
plt.suptitle("Freqncia de temps de retard a la sortida, a l'arribada i el temps en ruta")
plt.xticks(np.arange(len(df["UniqueCarrier"].unique())), df["UniqueCarrier"].unique())
plt.legend()
plt.show()
```
