```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
sns.set_context("paper", font_scale=1.5)

dades = sns.load_dataset("tips")

print(dades.isnull().sum())


sns.countplot(x = "time", data = dades).set(title='Reserves per tipus de menjar') 
plt.show()

sns.countplot(x="sex", data=dades).set(title='Reserves per sexe')
plt.show()

plt.figure()
sns.histplot(data = dades, x="total_bill").set(title='Distribució de les les factures') 
plt.show()

sns.regplot(x = "total_bill", y = "tip", data = dades)
plt.suptitle("Linearitat entre el que es gasta i la propina")
plt.show()

sns.boxplot(x="day", y="total_bill", data=dades, hue="sex").set(title='Factures per dia i sexe')
plt.legend(loc=0) 
plt.show()


```

```python
sns.set_theme(font_scale=1.2)

plt.figure(figsize=(14, 5))

sns.scatterplot(data=dades,
                x="total_bill",
                y="tip",
                hue="sex"
               )

plt.title("Relació entre factures i propina")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data= dades.groupby(["day", "sex"])["tip"].sum().reset_index(),
            x="day",
            y="tip",
            hue="sex",
            order = [ "Thur", "Fri", "Sat", "Sun"],
            palette = "magma"
           )

plt.title("Propina segons el dia i el sexe")
plt.show()
```

```python
sns.pairplot(data=dades,
             vars = ["tip", "total_bill", "size"], 
             kind="scatter",                      
             diag_kind="kde",                       
             hue="smoker",
             height=2,
             aspect=1.1,
             palette="bright")

plt.suptitle("Relacions entre factura, propina i la grandària de la reserva  \n segons si es fuma o no", y=1.08)   # adds title to the plot
plt.show()
```

```python
plt.figure(figsize=(14, 5))
sns.histplot(data=dades,
             x="total_bill",
             kde=True,          
             bins=25           
            )
plt.ylabel("")
plt.title("Distribució de les factures")
plt.show()
```

```python
plt.figure(figsize=(9, 6))
dades['factpax'] = dades['total_bill'] / dades['size']
sns.boxplot(x = "size", y = "factpax", data = dades)
plt.suptitle("Factura mitjana per persona segons la grandària de la reserva")
plt.ylabel("")
plt.show()
```

```python
sns.barplot(x = "sex", y = "porctip", hue = "smoker", data = dades)
plt.suptitle("Percentatge de la propina en funció del sexe i si es fuma")
plt.show()
```

```python
sns.barplot(x = "time", y = "tip", hue = "smoker", data = dades)
plt.suptitle("Propina segons el menjar i si es fuma")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x = "total_bill", y = "porctip", data = dades)
plt.suptitle("Linearitat entre el que es gasta i la propina")
plt.show()
```

```python
g = sns.FacetGrid(dades, col = "time")
g.map(plt.hist, "tip", bins =10)
plt.suptitle("Frecuencia de la propina per menjar", y =1.02)
plt.show()
```
