```python
import numpy as np
import pandas as pd

df = pd.read_csv("registro-clubes-deportivos.csv", sep = ";")
print(df.isnull().sum())
```

```python
df.drop(["Teléfono", "Domicilio", "Fax", "Email", "Web", "F.Fundación", "Column 14"], axis=1, inplace=True)
df = df[df['Deportes'].notna()]
df
```

```python
# Mida de la mostra 86
# Mostra simple aleatòria

df_aleat = df.sample(n=86)
df_aleat
```

```python
# Mida de la mostra 86
# Mostra sistemàtica
import random

mostra2 = []
p = random.randint(0,len(df)+1)

for i in range(1,87):
    mostra2.append(p)  
    if p > len(df)-100:
        p = 99 -(len(df) - p)
    else:
        p = p + 100

df_sistem = df.iloc[mostra2]
df_sistem    


```

```python
# Mida de la mostra 86
# Mostra estratificada

# grups per província
# trobar el número de elements per província per a la mostra
represt_grups = df["Provincia"].value_counts().sort_index().to_frame()

represt_grups_num = represt_grups * (85/8595)
print(represt_grups_num)

df2 = df.sort_values(by=['Provincia'],ignore_index=True)

# 1 manera
df_estrati2 = []
start = 0
tope = 0
for index, row in represt_grups.iterrows():
    tope += int(round(represt_grups.loc[index, 'Provincia'], 0))
    for ite in range(0, int(round(represt_grups_num.loc[index, 'Provincia'], 0))):
        p = random.randint(start, tope)
        df_estrati2.append(df2.iloc[p])                   
    start = tope
print(pd.DataFrame(df_estrati2))

# 2 manera
df_estrati = []
df_estrati = df.groupby('Provincia', group_keys=False).apply(lambda x: x.sample(frac=0.01)) 
df_estrati   


```
