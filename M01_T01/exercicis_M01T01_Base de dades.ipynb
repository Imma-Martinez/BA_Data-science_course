```python
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database = 'SGAE_Practica')

mycursor = mydb.cursor()

TABLES = {}
TABLES['tb_genre'] = (
    "CREATE TABLE `tb_genre` ("
    "  `genre_id` int(11) NOT NULL,"
    "  `genre_name` varchar(40) NOT NULL,"
    "  `created_by_user` varchar(10) NOT NULL DEFAULT 'OS_SGAD',"
    "  `created_date` date,"
    "  `updated_date` date,"
    "  CONSTRAINT `pk_genre` PRIMARY KEY (`genre_id`)"
    ") ENGINE=InnoDB")
    
TABLES['tb_movie'] = (
    "CREATE TABLE `tb_movie` ("
    "  `movie_id` int(11) NOT NULL,"
    "  `movie_title` varchar(100) NOT NULL,"
    "  `movie_date` date,"
    "  `movie_format` varchar(50),"
    "  `movie_genre_id` int(11),"   
    "  `created_by_user` varchar(10) NOT NULL DEFAULT 'OS_SGAD',"
    "  `created_date` date,"
    "  `updated_date` date,"
    "  CONSTRAINT `pk_movie` PRIMARY KEY (`movie_id`),"
    "  CONSTRAINT `fk_movie_genre` FOREIGN KEY (`movie_genre_id`) "
    "     REFERENCES `tb_genre` (`genre_id`)"
    ") ENGINE=InnoDB")

TABLES['tb_role'] = (
    "CREATE TABLE `tb_role` ("
    "  `role_id` int(11) NOT NULL,"
    "  `role_name` varchar(60) NOT NULL,"
    "  `created_by_user` varchar(10) NOT NULL DEFAULT 'OS_SGAD',"
    "  `created_date` date,"
    "  `updated_date` date,"
    "  CONSTRAINT `pk_role` PRIMARY KEY (`role_id`)"
    ") ENGINE=InnoDB")

TABLES['tb_person'] = (
    "CREATE TABLE `tb_person` ("
    "  `person_id` int(11) NOT NULL,"
    "  `person_name` varchar(100) NOT NULL,"
    "  `person_country` varchar(40)," 
    "  `person_dob` date NOT NULL,"
    "  `person_dod` date,"
    "  `person_parent_id` int(11),"
    "  `created_by_user` varchar(10) NOT NULL DEFAULT 'OS_SGAD',"
    "  `created_date` date,"
    "  `updated_date` date,"
    "  CONSTRAINT `pk_person` PRIMARY KEY (`person_id`),"
    "  CONSTRAINT `fk_person_parent` FOREIGN KEY (`person_parent_id`) " 
    "     REFERENCES `tb_person` (`person_id`)"
    ") ENGINE=InnoDB")

TABLES['tb_movie_person'] = (
    "CREATE TABLE `tb_movie_person` ("
    "  `movie_id` int(11) NOT NULL,"
    "  `person_id` int(11) NOT NULL,"
    "  `role_id` int(11) NOT NULL,"
    "  `movie_award_ind` char(1) NOT NULL,"
    "  `created_by_user` varchar(10) NOT NULL DEFAULT 'OS_SGAD',"
    "  `created_date` date,"
    "  `updated_date` date,"
    "  CONSTRAINT `pk_movper` PRIMARY KEY (`movie_id`, `person_id`, `role_id`),"
    "  CONSTRAINT `fk_movper_movie` FOREIGN KEY (`movie_id`) "
    "     REFERENCES `tb_movie` (`movie_id`),"
    "  CONSTRAINT `fk_movper_person` FOREIGN KEY (`person_id`)" 
    "     REFERENCES `tb_person` (`person_id`),"
    "  CONSTRAINT `fk_movper_role` FOREIGN KEY (`role_id`)" 
    "     REFERENCES `tb_role` (`role_id`)"
    ") ENGINE=InnoDB")


for tabla_nom in TABLES:
    tabla_des = TABLES[tabla_nom]
    mycursor.execute(tabla_des)
    print("Tabla creada {}: ".format(tabla_nom), end='')
```

Base de dades que es diu SGAE_Practica amb 5 taules:
- tb_genre on es llisten totes els gèneres amb els que es poden classificar les pel.lícules de la base de dades. Té columnes:
     - d'identificació que es la clau primària, tipus integral que no pot estar buida
     - del gènere de tipus string de 40 caràcters que no es pot estar buida
     - de la persona que ha creat el gènere de tipus string i 10 caràcters, que no es pot deixar buida i per defecte es posa 'OS_SGAD'
     - dos camps de dat de creació i d'actualització
     - relació amb taula tb_movie de 1:N
    
- tb_movie, on es llisten totes les pel-licules amb les columnes:
     - id integral no buit que es la clau primaria
     - el títol de tipus string no vuida de 100 caràcters          
     - data de la pel.licula
     - format de la pel.licula, tipus strin de 50 caràcters
     - el id gènere que es clau forania i que conecta amb la taula tb_genre amb el mateix camp 
      
- tb_role, que recull la funció que pot fer una persona dins la pel.licula. Columnes:
     - id integral no buit i clau primaria
     - nom de la funció no buit de tipus string de 60 caràcters
     - usuari que ha creat el roll, tipus string no buit de 10 caràcter i per defecte 'OS_SGAD'
     - data de creació
     - data d'actualització
     - relació amb taula tb_movie_person de 1:N
     
- tb_person on es llista els noms de les persones relacionades amb les pel.lícules. Columnes:
     - id, integral no buit i clau primària
     - nom de la persona, de tipus string no buit de 100 caràcters
     - país, tipus de string de 40 caràcters
     - dos camps de dates relacionades amb la persona, naixement no buida i mort
     - un segon id que pot ser buit i que és clau forani amb el id clau primari d'aquesta mateixa taula  (això és correcte???????)
     - camps de data de creació del registre i data d'actualització
     
- tb_movie_person on es creuen totes les películes que han fet les persones, els seus rolls i si ha tingut premi. Columnes:
     - id de la película, id de la pesona, id de la funció. Totes tres camps junts fan la clau primaria composta. 
     - els tres camps mencionats son claus foranis dels seus respectius claus primaris a les taules de tb_movie, tb_person i tb_role
     - si té premi, de tipus string d'un caràcter

```python
# Exercici 3

query = "SELECT `genre_name` AS GENERE, count(`movie_title`) AS NUM_PEL \
  FROM tb_genre AS a \
  JOIN tb_movie AS b ON a.`genre_id` = b.`movie_genre_id`\
  GROUP BY GENERE ORDER BY NUM_PEL DESC"

sumpelisgenre = pd.read_sql_query(query, con = mydb)

sumpelisgenre
```

```python
# exercici 4

query = "SELECT NOM, NUM_ROL \
    FROM (\
      SELECT `person_name` AS NOM, `movie_id`, count(`role_id`) AS NUM_ROL\
      FROM tb_person AS a \
      JOIN tb_movie_person AS b ON a.`person_id` = b.`person_id` \
      GROUP BY CONCAT(b.`person_id`, b.`movie_id`))temp \
    WHERE temp.NUM_ROL > 1"

nombrerolsxpeli = pd.read_sql_query(query, con = mydb)

nombrerolsxpeli
```
