```python
# www.bolsamadrid.es i BeautifulSoup

import requests
from bs4 import BeautifulSoup

URL = 'https://www.bolsamadrid.es/esp/aspx/Indices/Resumen.aspx'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
columnes = soup.find_all('th')
nomscol = []
for job_element in columnes:
    nomscol.append(job_element.text)
    print(job_element.text, end="\n")
```

```python
fila = soup.find_all('tr', {'align':'right'}, limit = 30)

lista = []
for job_element in fila:
    dades = job_element.find_all('td')
    lista1=[]
    for dada in dades:
        lista1.append(dada.text)
    lista.append(lista1)
```

```python
import pandas as pd

df = pd.DataFrame (lista, columns = nomscol)
df
```

```python
# quotes.toscrape.com i BeautifulSoup

URL1 = 'http://quotes.toscrape.com/'
page1 = requests.get(URL1)

soup1 = BeautifulSoup(page1.content, 'html.parser')
```

```python
categories = soup1.find_all("span", class_="tag-item")
categorysoup = []
for job_element in categories:
    if job_element != None:
        element = job_element.text
        categorysoup.append(element.strip())

menu_opcions = {}
for item in range (0, 3):
    menu_opcions[item] = categorysoup[item]
menu_opcions[3] ='sortir'

def menu():
    print('Les 3 categories amb més entrades son:')
    for key in menu_opcions.keys():
        print(key+1, '.-', menu_opcions[key])

def getquotes(opcio):
        soup2 = BeautifulSoup(requests.get('http://quotes.toscrape.com/tag/'+ menu_opcions[opcio]).content, 'html.parser')    
        quotes = soup2.find_all("div", class_="quote")
        for quote_element in quotes:
            print(quote_element.find("span", class_="text").text)
            print(quote_element.find("small", class_="author").text)
            print()    
        
while True:
    menu()
    opcio = int(input('Indica el número de la teva opció: '))
    print()
    if (opcio == 1) or (opcio ==2) or (opcio ==3):
        print('Las citas de la teva selecció son:')
        getquotes(opcio-1)
    elif opcio == 4:
        break
    else:
        print('Incorrecte, indica un núm del 1 al 4')
        print()
```
